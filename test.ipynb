```python
#import pandas
import pandas as pd
```

```python
#Import pickle and pickle file named customer_id_merged.pkl
import pickle
df = pickle.load(open("customer_id_merged.pkl", "rb"))

#Print the first 5 rows of the dataframe
print(df.head())
```

```python
#impor the standard scaler file called standard_scaler.pkl
from sklearn.preprocessing import StandardScaler
scaler = pickle.load(open("standard_scaler.pkl", "rb"))
```

```python
df.columns
```

```python
last_three_columns = ['demographics_labels','purchase_behavior_labels', 'merged_labels']

df_last_three = df[last_three_columns]
df_no_last_three = df.drop(columns=last_three_columns)


#Print the first 5 rows of the dataframe
print(df_no_last_three.head())
```

```python
# Store original column names before transformation
original_columns = df_no_last_three.columns

# Apply the scaler to the dataframe
df_no_last_three = pd.DataFrame(scaler.inverse_transform(df_no_last_three))

# Restore original column names
df_no_last_three.columns = original_columns
df_no_last_three.index = df.index
```

```python
df_no_last_three.head()
```

```python
#now i want to merge the df_no_last_three and df_last_three
df_merged = pd.concat([df_no_last_three, df_last_three], axis=1)
df_merged.head()
```

```python
#export the dataframe to a pickle file
df_merged.to_pickle("customer_id_merged_unscaled.pkl")
```
